```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array([2,4,6,8,10])
y=np.array([1,2,3,4,5])
plt.subplot(2,2,1)
plt.plot(x,y)
x=np.array([2,4,6,8,10])
y=np.array([5,10,15,20,25])
plt.subplot(4,2,2)
plt.plot(x,y)
plt.show()
```

```python
x=np.array([1,2,3,4])
y=np.array([1,2,3,4])
plt.plot(x,y)
plt.show()
```

```python
x=np.array([1,10])
y=np.array([1,10])
plt.xlabel("Number")
plt.ylabel("square")
plt.show()
```

```python
x=np.array([2,7])
y=np.array([10,20])
plt.plot(x,y,'*:y')
plt.show()
```

```python
plt.scatter(x,y)
plt.show()
```

```python
x=np.array([1,6])
y=np.array([10,40])
plt.xlabel("Number")
plt.ylabel("square")
plt.xlim(3)
plt.ylim(20,25)
plt.plot(x,y)
plt.legend("singleline")
plt.show()
```

```python
x=np.arange(1,2*np.pi,5)
s=np.sin(x)
c=np.cos(x)
plt.plot(s,x,c)
plt.legend(['sin(x)','cos(x)'])
plt.show()
```

```python
x=np.random.randint(100,size=(5))
plt.hist(x,density=100)
plt.show()
```

```python
data={'Sree': 94,'Praveena': 96,'Ayesha':85,'Anusha':80,'Shilpa':100,'Ray':30}
name=list(data.keys())
mark=list(data.values())
plt.bar(range(len(data)), mark, tick_label=name)
plt.title("Student report")
plt.xlabel("Name")
plt.ylabel("Mark")
plt.show()
```
